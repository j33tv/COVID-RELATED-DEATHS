```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load CSV file into a DataFrame
csv_file_path = r'C:\\Users\\jeet\\Downloads\\AH_Excess_Deaths_by_Sex__Age__and_Race_and_Hispanic_Origin.csv'  
data = pd.read_csv(csv_file_path)
data.head()
```

```python


# Filter data for weeks 1 to 34 and selected years (2018, 2019, 2020)
years_of_interest = [2018, 2019, 2020]

# Filter data for selected years and weeks 1 to 34
filtered_data = data[(data['MMWRyear'].isin(years_of_interest)) & (data['MMWRweek'].between(1, 34))]

# Grouping data by MMWRyear and calculating total deaths for each year
total_deaths_per_year = filtered_data.groupby('MMWRyear')['Deaths (unweighted)'].sum()

# Plotting total deaths for the identified years in a horizontal bar plot and labeling as Year A, Year B, Year C
plt.figure(figsize=(8, 6))
total_deaths_per_year.plot(kind='barh', color='skyblue')  # Horizontal bar plot

plt.xlabel('Total Deaths in USA in weeks 1-34')
plt.ylabel('Year')
plt.yticks(range(len(years_of_interest)), ['Year A', 'Year B', 'Year C'])  # Set custom labels for Y-axis
plt.show()
```
